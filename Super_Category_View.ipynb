- User Defined Functions

```python
# Advance Report
def fill_Neo(template_df, source_df, main_col, ad_group, sub_col, sum_col):
    if main_col == 'Alpha':
        mask = (
            ((source_df['New BU'] == '3P') & (source_df['New Alpha/MP'] == '3P')) |
            ((source_df['New BU'] != '3P') & (source_df['New Alpha/MP'] == 'Alpha'))
        )
    else:
        mask = (source_df['New Alpha/MP'] == main_col)
        
    filtered = source_df[mask]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    summary.index.names = template_df.index.names
    final_summary = summary.reindex(template_df.index).fillna(0)
    template_df.loc[:, (main_col, ad_group, sub_col)] = final_summary.values
    
    return template_df
```

```python
# Pipiline report
def fill_Non_Neo(template_df, source_df, main_col, ad_group, sub_col, sum_col):
    filtered = source_df[(source_df['New Alpha/MP'] == main_col) & (source_df['Rate Card Bifurcation'] == "Sponsorship")]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    summary.index.names = template_df.index.names
    final_summary = summary.reindex(template_df.index).fillna(0)
    template_df.loc[:, (main_col, ad_group, sub_col)] = final_summary.values
    
    return template_df
```

```python
# Credits
def fill_credits(template_df, source_df, main_col, ad_group, sub_col, sum_col):
    filtered = source_df[source_df['Alpha_MP'] == main_col]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    summary.index.names = template_df.index.names
    final_summary = summary.reindex(template_df.index).fillna(0)
    template_df.loc[:, (main_col, ad_group, sub_col)] = final_summary.values
    
    return template_df
```

```python
# Brand_PLA
def fill_brand_pla(template_df, source_df, main_col, ad_group, sub_col, sum_col):
    filtered = source_df[(source_df['Alpha/MP'] == main_col) & (source_df['Type'] == "BRAND_PLA")]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    summary.index.names = template_df.index.names
    final_summary = summary.reindex(template_df.index).fillna(0)
    template_df.loc[:, (main_col, ad_group, sub_col)] = final_summary.values
    
    return template_df
```

```python
# Brand_PCA
def fill_brand_pca(template_df, source_df, main_col, ad_group, sub_col, sum_col):
    filtered = source_df[(source_df['Alpha/MP'] == main_col) & (source_df['Campaign Type'] == "BRAND_PCA")]
    summary = filtered.groupby(['BU', 'Super category'])[sum_col].sum()
    summary.index.names = template_df.index.names
    final_summary = summary.reindex(template_df.index).fillna(0)
    template_df.loc[:, (main_col, ad_group, sub_col)] = final_summary.values
    
    return template_df
```

```python
# PLA
def fill_pla(template_df, source_df, main_col, ad_group, sub_col, sum_col):
    filtered = source_df[(source_df['Alpha/MP'] == main_col) & (source_df['Type'] == "PLA")]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    summary.index.names = template_df.index.names
    final_summary = summary.reindex(template_df.index).fillna(0)
    template_df.loc[:, (main_col, ad_group, sub_col)] = final_summary.values
    
    return template_df
```

```python
# PCA
def fill_pca(template_df, source_df, main_col, ad_group, sub_col, sum_col):
    filtered = source_df[(source_df['Alpha/MP'] == main_col) & (source_df['Type'] == "PCA")]
    summary = filtered.groupby(['New BU', 'New Supercat'])[sum_col].sum()
    summary.index.names = template_df.index.names
    final_summary = summary.reindex(template_df.index).fillna(0)
    template_df.loc[:, (main_col, ad_group, sub_col)] = final_summary.values
    return template_df
```

```python
# Paid Credits
def fill_paid_credits(template_df, source_df, main_col, ad_group, sub_col, sum_col):
    summary = source_df.groupby(['New BU','New Supercat'])[sum_col].sum()
    summary.index.names = template_df.index.names
    final_summary = summary.reindex(template_df.index).fillna(0)
    template_df.loc[:, (main_col, ad_group, sub_col)] = final_summary.values
    
    return template_df
```

```python
# 'Display'
def fill_display(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[(source_df['New Alpha/MP'] == main_col) & (source_df['Check'] == "Yes")]
    summary = filtered.groupby(['New BU','New Supercat'])[sum_col].sum()
    # summary.index.names = template_df.index.names
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0).values
        
    return template_df
```

```python
# 'Search'
def fill_search(template_df, source_df, main_col, sub_col, sum_col):
    filtered = source_df[source_df['Alpha_MP'] == main_col]
    summary = filtered.groupby(['New BU','New Supercategory'])[sum_col].sum()
    template_df.loc[:, (main_col, sub_col)] = summary.reindex(template_df.index).fillna(0)
    
    return template_df
```

- Exporting to Excel .xlsx format

```python
# 1. Setup the Excel Writer
file_path = "Super Catagory View.xlsx"
writer = pd.ExcelWriter(file_path, engine='xlsxwriter')
workbook  = writer.book
worksheet = workbook.add_worksheet('SuperCategoryView')

# 2. Define Custom Formats
title_format = workbook.add_format({
    'bold': True, 
    'align': 'center', 
    'font_size': 14, 
    'bg_color': '#4472C4', 
    'font_color': 'white',
    'border': 2
})
thin_format = workbook.add_format({'border': 1})
thick_border_val = 2

# 3. List of Tables and their Display Names
# Pairing the DataFrame with the Title we want to show
tables = [
    (df_consumption, "CONSUMPTION DATA"),
    (df_redemption, "REDEMPTION DATA"),
    (df_estimations, "ESTIMATION DATA"),
    (df_search_roins, "SEARCH ROINS")
]

current_col = 0
gap = 2 
title_row = 0  # Row index for the Title
start_row = 1  # Row index where the Table headers start

for df, title in tables:
    # --- A. Calculate Dimensions ---
    n_header_rows = df.columns.nlevels
    n_index_cols = df.index.nlevels
    total_cols = len(df.columns) + n_index_cols
    total_rows = len(df.index) + n_header_rows + 1 # +1 for title row
    
    # --- B. Write Title ---
    # Merge cells across the width of the table
    worksheet.merge_range(title_row, current_col, title_row, current_col + total_cols - 1, title, title_format)
    
    # --- C. Write DataFrame ---
    df.to_excel(writer, sheet_name='SuperCategoryView', startrow=start_row, startcol=current_col)
    
    # --- D. Apply Internal Thin Borders (Grid) ---
    worksheet.conditional_format(title_row, current_col, total_rows - 1, current_col + total_cols - 1,
                                 {'type': 'no_errors', 'format': thin_format})
    
    # --- E. Apply Thick Outer Borders ---
    # Top
    worksheet.conditional_format(title_row, current_col, title_row, current_col + total_cols - 1,
                                 {'type': 'no_errors', 'format': workbook.add_format({'top': thick_border_val})})
    # Bottom
    worksheet.conditional_format(total_rows - 1, current_col, total_rows - 1, current_col + total_cols - 1,
                                 {'type': 'no_errors', 'format': workbook.add_format({'bottom': thick_border_val})})
    # Left
    worksheet.conditional_format(title_row, current_col, total_rows - 1, current_col,
                                 {'type': 'no_errors', 'format': workbook.add_format({'left': thick_border_val})})
    # Right
    worksheet.conditional_format(title_row, current_col + total_cols - 1, total_rows - 1, current_col + total_cols - 1,
                                 {'type': 'no_errors', 'format': workbook.add_format({'right': thick_border_val})})

    # Move to the next start position
    current_col += total_cols + gap

# 4. Final Polish
worksheet.hide_gridlines(2)
worksheet.set_column(0, current_col, 15) # Set a readable column width

# 5. Save
writer.close()
```
